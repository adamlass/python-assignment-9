### Questions/Exercises:
1. Download the dataset programatically, and split the data into a train-test set.

```python
#Here i will use tensorflow to import the same dataset instead.
import tensorflow as tf
from tensorflow import keras

#loading dataset
fashion_mnist = keras.datasets.fashion_mnist

#test-train split
(train_images, train_labels), (test_images, test_labels) = fashion_mnist.load_data()

#printing out shapes
print("train_images: ", train_images.shape)
print("train_labels: ", train_labels.shape)

print("\ntest_images", test_images.shape)
print("test_labels", test_labels.shape)
```

```python
#Plotting to see the contents
import matplotlib.pyplot as plt

plt.figure()
plt.imshow(train_images[0])
plt.colorbar()
plt.gca().grid(False)
```

 2. Using TensorFlow, create a neural network and train it.

```python
#Declaring the corresponding class names
class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat', 
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']
```

```python
#Normalizing the greyscale values
train_images = train_images / 255.0
test_images = test_images / 255.0
```

```python
#Plotting some of the data
plt.figure(figsize=(10,5))
for i in range(15):
    plt.subplot(3,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.imshow(train_images[i])
    plt.xlabel(class_names[train_labels[i]])
plt.show()
```

```python
#Declaring model
model = keras.Sequential([
    keras.layers.Flatten(input_shape=(28, 28)),
    keras.layers.Dense(128, activation=tf.nn.relu),
    keras.layers.Dense(10, activation=tf.nn.softmax)
])
```

```python
#Setting optimizer, loss-function and accuracy metrics
model.compile(optimizer='adam', 
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
#training the model
model.fit(train_images,train_labels,epochs=5)
```

3. Use the provided test data, and evaluate the accuracy of the network.




```python
predictions = model.predict(test_images)
```

```python
#Looking at the very first prediction
import numpy as np

np.argmax(predictions[0])
```

```python
#Getting the actual value
test_labels[0]
```

> We can see that the model was accurate on the first dataset. But what about the total accuracy?

```python
test_loss, test_acc = model.evaluate(test_images, test_labels)

print(int(test_acc * 100), "% of the guesses are correct!")
```

> We can see here that this is a pretty good fit, meaning that the model is very accurate. This is probably also because of the images is very alike in style throughout the dataset, and maybe given some other types of images (downscaled to 28x28), it would probably perform worse that it does currently.
